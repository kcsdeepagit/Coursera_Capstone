# Capstone Project  - The Battle of Neighborhoods |Bengaluru Vs Hyderabad, India

# 1. Introduction

This project is to find a better place which has all the facilities in Bengaluru and Hyderabad, which is the IT capital of India . Lot of people come here for job and they face problems in finding a better place to live in , this project helps all those in making smart and efficient decision on selecting great neighborhood out of numbers of other neighborhoods in Bengaluru and Hyderabad , India.

People who come here for jobs need to do lots of research for good housing prices and reputed schools for 
their children. This project is for those people who are looking for better neighborhoods. 
For ease of accessing to Cafe, School, Super market, medical shops, Parks,grocery shops, mall, theatre, hospital, like minded people, etc.

This Capstone Project aim to create an analysis of features for a people migrating to Bengaluru to search a best neighborhood as a comparative analysis between neighborhoods. The features include median housing price and better school according to ratings, crime rates of that particular area, road connectivity, weather conditions, good management for emergency, water resources both fresh and waste water and recreational facilities.

It will help people to get awareness of the area and neighborhood before moving to a new city, state, country or place for their work or to start a new fresh life.

    Following are the steps need to be followed to find the better place:
    
    1.Scrape the data using beautifulsoup.
    2.Find the latitudes and longitudes of the given addresses(using Gmaps geocoding API)
    3.Plot those coordinates using the folium map.
    4.Find the number of collisions in the given dataset.
    5.Using the dataset, we will use FourSquare API for finding the places resided near to those pin codes and merge with the original       dataset
    6.To use this dataset for predicting the most common venue, we have to use one-hot encoding for K-means Clustering algorithm.
    7.Using the silhouette score, we will find the best K-value for our most common venue prediction.
    8.Using the K-value, we will find the cluster labels for each an every place and last we will plot the map and find the most
      common venue point in the selective district.
